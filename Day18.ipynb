```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlt
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
profit=df.total_profit
month = df.month_number
```

```python
month
```

```python
profit
```

```python
plt.plot(month,profit)
plt.xlabel(r'Month number')
plt.ylabel(r'Total profit')
plt.title(r'Company profit per month')
plt.xticks(month)
plt.yticks([100000,200000,300000,400000,500000])
```

```python
plt.plot(month,profit,label='profit data of last year',color='r',marker='o',linestyle='--',linewidth=3,markerfacecolor='k')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(month)
plt.yticks([100000,200000,300000,400000,500000])
plt.show()
```
