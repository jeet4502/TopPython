```python
class A:
    def addition(x,y):
        return x+y
class B:
    def subtraction(x,y):
        return x-y
class C:
    def multiplication(x,y):
        return x*y
class D(A,B,C):
    def result(self):
        addition()
        subtraction()
        multiplication()
a = D
print(a.addition(10,5))
print(a.subtraction(10,5))
print(a.multiplication(10,5))
```

```python
'''
Different forms
Operator level Polymorphism: 2+3,'abc,'def'
Function level Polymorphism:len([1,2,3]),len("abcdef"),len(1,2,3,4)
Class level Polymorphism
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mlt
plt.figure(figsize=(10,10))
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.bar(id1,salary)
plt.xlabel("Employee id")
plt.ylabel('salary')
plt.show()
```

```python
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.figure(figsize=(8,5))
ax = plt.axes()
ax.set_facecolor("#F4D03F")
plt.bar(id1,salary,color="#F16F15")
plt.xlabel(r'$Employee $$ id$')
plt.ylabel(r'$ Salary $')
plt.show()
           
```
