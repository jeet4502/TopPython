```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
titanic=sns.load_dataset('titanic')
planets=sns.load_dataset('planets')
```

```python
tips
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue="day",size="size",palette="YlGnBu")
#hue, showed tips given by day in different colors
#size, changed the size of the dots based on the amount of tips given by the day
#palette, allowed us to change the color of the dots
```

```python
sns.histplot(tips['tip'],kde=True,bins=15)
#kde enabled the line, if we do 'False' the line will go
#bins, number of boxes/bars in the histogram
```

```python
sns.displot(tips['tip'],kde=True,bins=15)
```

```python
sns.barplot(x="sex",y="tip",data=tips,palette='YlGnBu')
```

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu")
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue="sex",palette="YlGnBu")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue='sex',dodge=True,palette="YlGnBu")
#When i set dodge=True it separated the dots of the gendeers. You can see the difference below when we do not use dodge
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue='sex',palette="YlGnBu")
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True)
#you can use fill=True instead of using shad=True but shade=True is more preferable
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True,cmap='YlGnBu')
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex",cmap="YlGnBu")
#kind="hex" is for hexagons
```
